Given a n x n matrix where each of the rows and columns are sorted in ascending order, find the kth smallest element in the matrix.

Note that it is the kth smallest element in the sorted order, not the kth distinct element.

Example:

matrix = [
             [ 1,  5,  9],
             [10, 11, 13],
             [12, 13, 15]
         ],
k = 8,

return 13.
Note:
You may assume k is always valid, 1 ≤ k ≤ n2.

```python
from collections import deque
class TreeNode(object):
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
class Solution(object):
    def kthSmallest(self, matrix, k):
        """
        :type matrix: List[List[int]]
        :type k: int
        :rtype: int
        """ 
        matrix_list = []
        
        n = len(matrix)
        d = 0
        
        while d < n:
            i = 0
            for i in range(0,d):
                matrix_list.append(matrix[i][d])
                matrix_list.append(matrix[d][i])
            matrix_list.append(matrix[d][d])
            d += 1
        head = self.buildHeap(matrix_list)
        result = head.val
        for i in range(0,k):
            head,last_parent,last_node, left = self.minHeapify(head)
            result = head.val
            if last_node is None:
                break
            last_node.val, head.val = head.val,last_node.val
            if left:
                last_parent.left = None
            else:
                last_parent.right = None
        return result
        
    def buildHeap(self,nums):
        head = TreeNode(nums[0])
        current = head
        node_list = deque()
        for i in nums[1:]:
            if current.left is None:
                current.left = TreeNode(i)
                node_list.append(current.left)
            elif current.right is None:
                current.right = TreeNode(i)
                node_list.append(current.right)
            else:
                current = node_list.popleft()
                current.left = TreeNode(i)
                node_list.append(current.left)
        return head
    def printHeap(self,head):
        node_list = [head,None]
        result = []
        if head is None:
            return head
        while len(node_list) != 0:
            current = node_list[0]
            if current is not None:
                result.append(current.val)
                if current.left is not None:
                    node_list.append(current.left)
                if current.right is not None:
                    node_list.append(current.right)
            else:
                print(result)
                result = []
                if len(node_list) == 1:
                    break
                node_list.append(None)
            node_list = node_list[1:]
    def minHeapify(self,head):
        parent_nodes = []
        travel = [head]
        last_parent = None
        last_node = None
        left = None
        while len(travel) != 0:
            current = travel.pop()
            if current.left is not None or current.right is not None:
                parent_nodes.append(current)
                if current.left is not None:
                    last_parent = current
                    last_node = current.left
                    left = True
                    travel.append(current.left)
                if current.right is not None:
                    last_parent = current
                    last_node = current.right
                    left = False
                    travel.append(current.right)
        while len(parent_nodes) != 0:
            current = parent_nodes.pop()
            if current.left is not None:
                if current.left.val < current.val:
                    current.val, current.left.val = current.left.val, current.val
            if current.right is not None:
                if current.right.val < current.val:
                    current.val, current.right.val = current.right.val, current.val
        return head, last_parent, last_node, left
        
        
            
        
            
            
        
        
        
```

```python
from heapq import heappush, heappop
class TreeNode(object):
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
class Solution(object):
    def kthSmallest(self, matrix, k):
        heap = []

        n = len(matrix)
        d = 0

        while d < n:
            for i in range(0,d):
                heappush(heap,matrix[i][d])
                heappush(heap,matrix[d][i])
            heappush(heap,matrix[d][d])
            d += 1
        for i in range(0,k):
            smallest = heappop(heap)
        return smallest 
            
        

```
